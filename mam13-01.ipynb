```python
import pandas as pd
```

```python
df = pd.read_csv("customer_encoder.csv")
df.info()
df
```

```python
# encoders
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder, OneHotEncoder
```

```python
df
```

```python
# ordinal encoder
x = df.iloc[:,1:3]
le = OrdinalEncoder(categories=[["Poor","Average","Good"],["School","UG","PG"]])
xx = le.fit_transform(x)
xx
```

```python
# label encoder
lle = LabelEncoder()
lle.fit_transform(df["purchased"])
```

```python
# onehotencoder
y = df.iloc[:,-1].values.reshape(-1,1)
ohe = OneHotEncoder()
xx = ohe.fit_transform(y).toarray()
```
