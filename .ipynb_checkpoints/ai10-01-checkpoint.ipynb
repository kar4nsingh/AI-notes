```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
import joblib
```

```python
df = pd.read_csv("50_startups.csv")
df
```

```python
da = df.drop(columns=["State"])
```

```python
da.info()
```

```python
df
```

```python
da
```

```python
sns.lineplot(da)
```

```python
x = da.iloc[:,:3]
x
```

```python
y = da.iloc[:,-1]
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 33)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
predict = lr.predict(x_test)
```

```python
da.head(3)
```

```python
#y = b1x1 + b2x2 + b3x3 + bo
bo = lr.intercept_
# print(bo)
bm = lr.coef_
# print(bm)
b1 = bm[0]
b2 = bm[1]
b3 = bm[2]
y = (b1 * 153441.51) + (b2 * 101145.55) + (b3 * 407934.54) + bo
y
```

```python
r2_score(y_test,predict)
```

```python
joblib.dump(lr,"multipleLR.joblib")
```
