```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
le = LabelEncoder()
```

```python
x = daa.iloc[:,-1]
print(x)
```

```python
#label_encoder
le.fit_transform(x)
```

```python
#onehot_encoder
y = daa.iloc[:,0].values.reshape(-1,1)
ohe = OneHotEncoder()
ohe.fit_transform(y).toarray()
```
